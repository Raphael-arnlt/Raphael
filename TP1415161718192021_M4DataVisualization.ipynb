# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
conn = sqlite3.connect('/resources/labs/module 4/m4_survey_data.sqlite')
df = pd.read_sql_query("SELECT * FROM master", conn)
conn.close()

df = df.dropna(subset=['ConvertedComp'])

plt.figure(figsize=(10, 6))
plt.hist(df['ConvertedComp'], bins=30, color='blue', alpha=0.7)
plt.title('Histogram of Converted Annual Salaries (USD)')
plt.xlabel('Converted Salary (USD)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
plt.figure(figsize=(10, 6))
plt.boxplot(df['Age'], vert=False)
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.grid(True)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['WorkWeekHrs'], alpha=0.7, color='blue')
plt.title('Scatter Plot of Age vs Work Week Hours')
plt.xlabel('Age')
plt.ylabel('Work Week Hours')
plt.grid(True)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
df = df.dropna(subset=['WorkWeekHrs', 'CodeRevHrs', 'Age'])

plt.figure(figsize=(12, 6))
bubble_size = df['Age'] * 10 
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=bubble_size, alpha=0.5, color='blue')
plt.title('Bubble Plot of Work Week Hours vs Code Review Hours')
plt.xlabel('Work Week Hours')
plt.ylabel('Code Review Hours')
plt.grid(True)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
top_experience = df['YearsCodePro'].value_counts().nlargest(5)

plt.figure(figsize=(10, 6))
plt.pie(top_experience, labels=top_experience.index, autopct='%1.1f%%', startangle=140)
plt.title('The 5 Programming Experience Levels Respondents Have')
plt.axis('equal')  
plt.show()
#XXX pas certain de la donnée
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
age_group = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

median_workweekhrs = age_group['WorkWeekHrs'].median()
median_coderevhrs = age_group['CodeRevHrs'].median()
data = pd.DataFrame({
    'Hours': [median_workweekhrs, median_coderevhrs],
    'Type': ['WorkWeekHrs', 'CodeRevHrs']
})

# Plot the stacked bar chart
plt.figure(figsize=(10, 6))
plt.bar(data['Type'], data['Hours'], color=['blue', 'orange'])
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
plt.ylabel('Hours')
plt.show()

```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
bins = [17, 24, 34, 44, 60, 100]
labels = ['18-24', '25-34', '35-44', '45-60', '60+']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)

# Calculer la médiane pour chaque groupe d'âge
median_converted_comp = df.groupby('AgeGroup')['ConvertedComp'].median()

# Tracer les médianes dans un graphique linéaire
plt.figure(figsize=(10, 6))
plt.plot(median_converted_comp.index, median_converted_comp.values, marker='o', linestyle='-', color='blue')
plt.title('Median Converted Compensation by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Median Converted Compensation (USD)')
plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
main_branch_counts = df['MainBranch'].value_counts()

plt.figure(figsize=(10, 6))
plt.barh(main_branch_counts.index, main_branch_counts.values, color='skyblue')
plt.title('Distribution of Main Branches')
plt.xlabel('Number of Respondents')
plt.ylabel('Main Branch')
plt.show()
```
